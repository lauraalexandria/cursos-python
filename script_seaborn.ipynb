# Seaborn

#### Importando bibliotecas

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/lauraalexandria/cursos_python/main/data/seattle_weather.csv")
```

#### Gr√°ficos de Pontos

```python
sns.scatterplot(data = df, x = "MLY-TAVG-NORMAL", y = "MLY-PRCP-NORMAL") 
plt.show()
```

```python
sns.scatterplot(data = df, x = "MLY-TAVG-NORMAL", y = "MLY-PRCP-NORMAL", hue = "MLY-PRCP-NORMAL") # Legenda com as cores pela categ√≥rica;
# hue_order ainda ajuda a ordenar na legenda e d√° pra definir em palette = {"Yes": "red", "No": "black"}
plt.show()
```

```python
sns.relplot(data = df, x = "MLY-TAVG-NORMAL", y = "MLY-PRCP-NORMAL", kind = "scatter", col = "DATE") 
# V√°rios gr√°ficos de pontos pelas classes da categoria, um do lado do outro
plt.show()
```

```python
sns.relplot(data = df, x = "MLY-TAVG-NORMAL", y = "MLY-PRCP-NORMAL", kind = "scatter", row = "DATE") 
# Vers√£o dos gr√°ficos em uma √∫nica linha;
plt.show()
```

```python
sns.relplot(data = df, x = "MLY-TAVG-NORMAL", y = "MLY-PRCP-NORMAL", kind = "scatter", col = "DATE", col_wrap = 6) 
plt.show()
```

```python
iris = sns.load_dataset("iris")
iris_long = pd.melt(iris, "species", var_name="measurement")
g = sns.FacetGrid(iris_long, hue="species", col="measurement", col_wrap=2, sharex=False)
g.map(plt.hist, "value", alpha=.4)
```

```python
sns.scatterplot(data = df, x = "MLY-TAVG-NORMAL", y = "MLY-PRCP-NORMAL", size = "DATE") # Permite que os pontos mudem de tamanho;
plt.show()
```

```python
sns.scatterplot(data = df, x = "MLY-TAVG-NORMAL", y = "MLY-PRCP-NORMAL", style = "DATE") # Permite que os pontos mudem de formato;
plt.show()
```

```python
sns.scatterplot(data = df, x = "MLY-TAVG-NORMAL", y = "MLY-PRCP-NORMAL", alpha = 0.2) 
# Altera a opacidade dos pontos, o que pode ser interessante quando √© sobreposi√ß√£o (alpha recebe valores entre 0 a 1); 
plt.show()
```

```python
sns.stripplot(data=df, x='DATE', y='MLY-TAVG-NORMAL', jitter=True)
# Quando uma das vari√°veis √© categ√≥rica;
plt.show()
```

#### Gr√°ficos de linhas 

```python
sns.relplot(data = df, x = "DATE", y = "MLY-PRCP-NORMAL", kind = "line")
# Se houverem diferentes observa√ß√£o para os valores de x, automaticamente temos um intervalo de confian√ßa para a linha;
plt.show()
```

```python
sns.relplot( data = df, x = "MLY-TAVG-NORMAL", y = "MLY-PRCP-NORMAL", kind = "line", style = "DATE", hue = "DATE", 
            markers = True, dashes = False ) # Linhas de acordo com as categorias, mostra os pontos tamb√©m e sem grades;
plt.show()
```

```python
sns.relplot(data = df, x = "DATE", y = "MLY-PRCP-NORMAL", kind = "line", ci = "sd")
# O intervalo de confian√ßa √© realizado com o desvio padr√£o;
plt.show()
```

```python
sns.relplot(data = df, x = "DATE", y = "MLY-PRCP-NORMAL", kind = "line", ci = None)
# Retira  o intervalo de confian√ßa;
plt.show()
```

```python
sns.catplot(x = "DATE", y = "MLY-PRCP-NORMAL", data = df, kind = "point") # Gr√°fico de linhas para vari√°veis categ√≥ricas!
plt.show()
```

```python
sns.catplot(x = "DATE", y = "MLY-PRCP-NORMAL", data = df, kind = "point", hue = "NAME") # Para ir separando as linhas
plt.show()
```

```python
sns.catplot(x = "DATE", y = "MLY-PRCP-NORMAL", data = df, kind = "point", hue = "NAME", join = False) # Fica s√≥ os pontos
plt.show()
```

#### Gr√°ficos de barras

```python
sns.countplot(data = df, x = "DATE") # Gr√°fico de Barras de frequ√™ncia;
plt.show()
```

```python
sns.countplot(y = df["DATE"]) # Barras horizontais;
plt.show()
```

```python
sns.catplot(data = df, x = "DATE", kind = "count") # Com outra fun√ß√£o, que tem args muito parecidos com o relplot!
plt.show() 
```

```python
# sns.catplot(x = "categ√≥rica", data = df, kind = "count",
#             order = ["cat1", "cat2", ...]) # Ordenando o as barras;
# plt.show() 
```

```python
cross_tab_prop = pd.crosstab(columns=df["DATE"],
                             index=df["NAME"],
                             normalize="index")#.sort_values("COluna para ordenar as barras")
cross_tab_prop.plot.barh(stacked=True, figsize=(12, 5))
```

```python
sns.catplot(x = "DATE", y = "MLY-PRCP-NORMAL", data = df, kind = "bar") # Barras com a m√©dia da vari√°vel pelas categorias, com um intervalo;
plt.show() 
```

#### Boxplot

```python
sns.catplot(x = "DATE", y = "MLY-PRCP-NORMAL", data = df, kind = "box") 
plt.show() 
```

```python
sns.catplot(x = "DATE", y = "MLY-PRCP-NORMAL", data = df, kind = "box", sym = "") # n√£o mostra os outlier
plt.show() 
```

```python
sns.catplot(x = "DATE", y = "MLY-PRCP-NORMAL", data = df, kind = "box", whis = 2.0) # muda a extens√£o das retas 
plt.show() 
```

```python
sns.catplot(x = "DATE", y = "MLY-PRCP-NORMAL", data = df, kind = "box", whis = [2, 95]) # muda a extens√£o das retas - em % 
plt.show() 
```

```python
sns.boxplot(data=df, x= 'DATE', y= 'MLY-PRCP-NORMAL') # Outra f√≥rmula;
plt.show()
```

#### Gr√°fico de Distribui√ß√£o

```python
sns.distplot(df['MLY-PRCP-NORMAL'], rug=True, kde_kws={'shade':True})
plt.show()
```

```python
sns.distplot(df['MLY-PRCP-NORMAL'], hist=False, rug=True, kde_kws={'shade':True})
plt.show()
```

#### Regress√£o

```python
sns.regplot(x = "MLY-PRCP-NORMAL", y = "MLY-TAVG-NORMAL", data = df) 
plt.show()
```

```python
sns.lmplot(data=df, x = "MLY-PRCP-NORMAL", y = "MLY-TAVG-NORMAL", col = "DATE", col_wrap = 6)   # Tamb√©m √© poss√≠vel separar por classes...
plt.show()
```

```python
sns.regplot(x = "MLY-PRCP-NORMAL", y = "MLY-TAVG-NORMAL", data = df, order = 2) # Alterando o grau do polin√¥mio 
plt.show()
```

```python
sns.residplot(x = "MLY-PRCP-NORMAL", y = "MLY-TAVG-NORMAL", data = df, color = "g") # An√°lise de Res√≠duos
plt.show()
```
