# Matplotlib

#### Importando bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

#### Lendo bancos de dados

```python
austin_weather = pd.read_csv("https://raw.githubusercontent.com/lauraalexandria/cursos_python/main/data/austin_weather.csv")
seattle_weather = pd.read_csv("https://raw.githubusercontent.com/lauraalexandria/cursos_python/main/data/seattle_weather.csv")
seattle_weather = seattle_weather.loc[range(0,12)]
climate_change = pd.read_csv("https://raw.githubusercontent.com/lauraalexandria/cursos_python/main/data/climate_change.csv")
```

#### Estrutura básica

```python
fig, ax = plt.subplots() # Criando os Eixos
plt.show() # E manda imprimir
```

```python
fig, ax = plt.subplots() 
ax.plot(seattle_weather["DATE"], seattle_weather['MLY-PRCP-NORMAL']) # Criando uma linha
ax.plot(austin_weather["DATE"], austin_weather['MLY-PRCP-NORMAL']) # Criando outra

plt.show() # Nunca esquecer
```

##### Costumizando o gráfico

```python
fig, ax = plt.subplots() 
ax.plot(seattle_weather["DATE"], seattle_weather['MLY-PRCP-NORMAL'], label = "Seattle") # Argumentos para definir a legenda;
ax.plot(austin_weather["DATE"], austin_weather['MLY-PRCP-NORMAL'], label = "Austin")

ax.legend() # Adicionando legenda
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(seattle_weather["DATE"], seattle_weather["MLY-PRCP-NORMAL"], marker = "o", color = "b", linestyle = "--") # Adicionando marcadores
ax.plot(austin_weather["DATE"], austin_weather["MLY-PRCP-NORMAL"], marker = "v", color = "r", linestyle = "--")

# Alterando os títulos
ax.set_xlabel("Time (months)")  
ax.set_ylabel("Precipitation (inches)")
ax.set_title("Weather patterns in Austin and Seattle")

plt.show()
```

##### Grid de gráficos

```python
fig, ax = plt.subplots(2, 2)

ax[0, 0].plot(seattle_weather["DATE"], seattle_weather["MLY-PRCP-NORMAL"])
ax[0, 1].plot(seattle_weather["DATE"], seattle_weather["MLY-TAVG-NORMAL"])
ax[1,0].plot(austin_weather["DATE"], austin_weather["MLY-PRCP-NORMAL"])
ax[1,1].plot(austin_weather["DATE"], austin_weather["MLY-TAVG-NORMAL"])

plt.show()
```

```python
fig, ax = plt.subplots(2, 1, sharey=True) # Faz com que os eixos sejam iguais

ax[0].plot(seattle_weather["DATE"], seattle_weather["MLY-PRCP-NORMAL"], color = "b")
ax[0].plot(seattle_weather["DATE"], seattle_weather["MLY-PRCP-25PCTL"], color = "b", linestyle = "--")
ax[0].plot(seattle_weather["DATE"], seattle_weather["MLY-PRCP-75PCTL"], color = "b", linestyle = "--")

ax[1].plot(austin_weather["DATE"], austin_weather["MLY-PRCP-NORMAL"], color = "r")
ax[1].plot(austin_weather["DATE"], austin_weather["MLY-PRCP-25PCTL"], color = "r", linestyle = "--")
ax[1].plot(austin_weather["DATE"], austin_weather["MLY-PRCP-75PCTL"], color = "r", linestyle = "--")

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(climate_change.index, climate_change["relative_temp"]) # Trabalhando com séries temporais;
ax.set_xlabel("Time")
ax.set_ylabel("Relative temperature (Celsius)")

plt.show()
```

```python
fig, ax = plt.subplots() # Adicionando anotações

ax.plot(climate_change.index, climate_change["relative_temp"])
ax.annotate(">1 degree", (500, 1))

plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(climate_change.index, climate_change["co2"], color="b") # Adicionando segundo eixo y

ax2 = ax.twinx()
ax2.plot(climate_change.index, climate_change["relative_temp"], color="r")

plt.show()
```

#### Tipos de Gráficos

##### Gráficos de Barras

```python
fig, ax = plt.subplots()
ax.bar(seattle_weather["DATE"], seattle_weather["MLY-PRCP-NORMAL"])
ax.set_ylabel("Nome do Eixo Y")
plt.show()
```

```python
fig, ax = plt.subplots() # Gráfico de barras empilhadas;
ax.bar(seattle_weather["DATE"], seattle_weather["MLY-PRCP-NORMAL"])
ax.bar(seattle_weather["DATE"], seattle_weather["MLY-PRCP-25PCTL"], bottom = seattle_weather["DATE"]) 
plt.show()
```

```python
fig, ax = plt.subplots() # Comparando informações de diferentes data-frames;
ax.bar("Categoria 1", seattle_weather["MLY-PRCP-NORMAL"].mean())
ax.bar("Categoria 2", austin_weather["MLY-PRCP-NORMAL"].mean()) 
plt.show()
```

##### Histogramas

```python
fig, ax = plt.subplots()
ax.hist(seattle_weather["MLY-PRCP-NORMAL"], bins = 10) # Mudando o número de barras;
plt.show()
```

```python
fig, ax = plt.subplots() # Juntando dois histogramas;
ax.hist(seattle_weather["MLY-PRCP-NORMAL"], bins = 10) 
ax.hist(austin_weather["MLY-PRCP-NORMAL"], bins = 10) 
plt.show()
```

##### Intervalos

```python
seattle_weather["VAR"] = seattle_weather["MLY-PRCP-NORMAL"].std()

fig, ax = plt.subplots() 
ax.errorbar(seattle_weather["DATE"], seattle_weather["MLY-PRCP-NORMAL"], yerr = seattle_weather["VAR"]) # O argumento yerr pode ser adicionado a outros graf.
plt.show()
```

##### Boxplot

```python
fig, ax = plt.subplots() 
ax.boxplot([seattle_weather["DATE"], seattle_weather["MLY-PRCP-NORMAL"]])
plt.show()
```

##### Gráfico de Pontos

```python
fig, ax = plt.subplots() 
ax.scatter(seattle_weather["MLY-TAVG-NORMAL"], seattle_weather["MLY-PRCP-NORMAL"])
plt.show()
```

```python
fig, ax = plt.subplots() # Uma forma de colorir segundo outra variável;
ax.scatter(seattle_weather["MLY-TAVG-NORMAL"], seattle_weather["MLY-PRCP-NORMAL"], c = seattle_weather["MLY-TMIN-NORMAL"])
plt.show()
```
